Importing all the relevant variables

```python
import numpy as np
import matplotlib.pyplot as plt
```

The first for loop is the number of times that we repeat the experience. I decided that we are going to repeat it 10 times.
The second for loop is the number of times that the person is going to throw the die.
The list 'random_walk' is the level of the person after each step, you can therefore understand that there will be 500 steps in this list.
We need to keep in mind that there's a 0.1% chance that the person will fall on each step, so there's an if statement whose role is to reset the step value to 0.

We then use the random module of the numpy library to throw the die and the conditions matches the problem. At the end, we will append the random_walk list to another list called all_walks. 
Since we are repeating the experience 10 times, there will be 10 lists of 500 steps each in the all_walks list.

```python
all_walks = []
for i in range(100):
    random_walk = [0]
    for x in range(500):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step= max(0,step-1)
        elif dice <=5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001:
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)
```

Lets transform the all_walks list into a 2D array (a list of lists)
In order to plot the results, we need to transpose this array. Instead of having 10 lists of 500 items, we will have 500 lists of 10 items.

The x-axis is number of steps taken.
The y-axis is the level of the person.

So in the data frame, the first row is the x-axis and the values of the items in the first row is the level of the person.

```python
np_data = np.array(all_walks)
np_data_t = np.transpose(np_data)
print(np_data_t)

plt.plot(np_data_t)
plt.xlabel('Number of steps')
plt.ylabel('Level of the person')
plt.title('The level of the person at each step using the die game')
plt.show()
```

In the last Data Frame, we will only look at the last row containing the last heights of each of the 100 trys. We will then put the datas into a histogram.

```python
last_height = np_data_t[-1,:]
plt.hist(last_height,bins = 15,edgecolor = 'black')
plt.xlabel('Level of the person after 500 steps')
plt.ylabel('Probability (%)')
plt.title('Probability of ending on a level playing the die game')
plt.show()
```

As you can see, we can have a rough idea of the level of

```python
probability = np.mean(last_height>=60)
print(f'The probability of ending up at level 60 or plus is roughly {probability*100}%')
```
